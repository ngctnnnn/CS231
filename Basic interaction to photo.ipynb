# Section 1: Đọc, hiển thị ảnh và dò tìm cạnh trên ảnh


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import os
dir_monhoc = '/gdrive/MyDrive/CS231.L22.KHCL'
```

```python
dir_baitap = os.path.join(dir_monhoc, 'EdgeDetection')
```

```python
#Đọc ảnh
import cv2 as cv 
path_filename = os.path.join(dir_baitap, 'peppers.jpg')
img = cv.imread(path_filename, 0)
```

```python
#Hiển thị ảnh 
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
#Nhân tích chập
kernels_gx = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]
                   ])
kernels_gy = np.array([[-1, -2, -1],
                       [0, 0, 0],
                       [1, 2, 1]
                       ]
                      )

gx1 = cv.filter2D(img, -1, kernels_gx)
gx2 = cv.filter2D(img, -1, -kernels_gx)
gy1 = cv.filter2D(img, -1, kernels_gy)
gy2 = cv.filter2D(img, -1, -kernels_gy)
G = np.sqrt(np.square(gx1 * 1.0)  + np.square(gx2 * 1.0) 
   + np.square(gy1 * 1.0) + np.square(gy2 * 1.0))
```

```python
#Dò tìm cạnh của ảnh
cv2_imshow(G)
```

```python
#Show plots
images = [img, gx1, gx2, gy1, gy2, G]

width, height = 255, 255
rows, cols = 3, 3

axes = []
fig = plt.figure()

for i in range(len(images)):
    image = images[i]
    axes.append(fig.add_subplot(rows, cols, i+1) )
    plt.imshow(image)

fig.tight_layout()    
plt.show()
```

#Section 2: Bộ lọc làm trơn ảnh
